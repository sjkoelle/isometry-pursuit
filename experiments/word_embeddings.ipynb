```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import torch
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sentence_transformers import SentenceTransformer
from scipy.spatial.distance import cosine
import os
os.environ["OMP_NUM_THREADS"] = "1"
os.environ["OPENBLAS_NUM_THREADS"] = "1"

from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt
import random
from sentence_transformers import SentenceTransformer


import os, numpy as np, pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import StandardScaler

# -------------------------------------------------------------
# 1.  vocabulary (6 classes, each 10 words) -------------------
word_categories = {
    "man": [
        "man", "male", "guy", "gentleman", "sir",
        #"lad", "chap", "bloke", "fellow", "husband"
    ],
    "woman": [
        "woman", "female", "lady", "madam", "gal",
        #"lass", "girl", "wife", "dame", "ma'am"
    ],
    "regent": [          # high‑status / ruler
        "regent", "sovereign", "monarch", "emperor", "pharaoh",
        #"tsar", "tsarina",# "kaiser", "sultan", "shah", "overlord"
    ],
    "slave": [           # low‑status / servitude
        "slave", "wench",#"serf", "thrall", "chattel",
        #"servant", "vassal", "indentured", "captive"
    ],
    "young": [
        "young", #"youth", "boy", "girl", "toddler",
        "kid",# "infant", "teenager", "adolescent", "juvenile"
    ],
    "old": [
        "old", "elderly",# "senior", "aged", "ancient",
        #"venerable", "geriatric", "elder", "oldster", "pensioner"
    ]
}

words  = sum(word_categories.values(), [])                       # flat list (N = 60)
word2i = {w: i for i, w in enumerate(words)}

# -------------------------------------------------------------
# 2.  embeddings ----------------------------------------------
model   = SentenceTransformer("all-MiniLM-L6-v2")
X       = model.encode(words, convert_to_tensor=True).cpu().numpy()   # (60, 384)
X_std = X

# -------------------------------------------------------------
# 3.  one‑dimensional axes by “centroid difference” ----------
pairs  = [("man", "woman"), ("regent", "slave"), ("young", "old")]
coords = np.zeros((len(words), len(pairs)))

for dim, (cls_pos, cls_neg) in enumerate(pairs):
    idx_pos = [word2i[w] for w in word_categories[cls_pos]]
    idx_neg = [word2i[w] for w in word_categories[cls_neg]]

    μ_pos = X_std[idx_pos].mean(axis=0)         # class‑centroid (positive)
    μ_neg = X_std[idx_neg].mean(axis=0)         # class‑centroid (negative)
    w     = μ_pos - μ_neg                       # direction vector

    # normalise to unit length (avoid divide‑by‑zero if the centroids coincide)
    w_norm = np.linalg.norm(w)
    if w_norm == 0:
        raise ValueError(f"Centroids for {cls_pos} vs {cls_neg} coincide!")
    w /= w_norm

    # project *all* words onto this axis
    coords[:, dim] = X_std @ w


axes          = ["gender", "status", "age"]
centroid_embedding = pd.DataFrame(coords, index=words, columns=axes)

print(centroid_embedding.head())
centroid_embedding  = StandardScaler().fit_transform(centroid_embedding)   
centroid_embedding = pd.DataFrame(centroid_embedding, columns = ['gender','status','age'], index =words)
# zero‑mean / unit‑var
print(centroid_embedding.head())

import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------------------------------------------
# 1.  add a category label for each word (once, right after building lda_embedding)
category_of = {w: cat for cat, vocab in word_categories.items() for w in vocab}
centroid_embedding["category"] = centroid_embedding.index.map(category_of).astype("category")

# -----------------------------------------------------------------
# 2.  seaborn pair‑plot coloured by category, with word labels
g = sns.pairplot(
    centroid_embedding,
    vars=["gender", "status", "age"],
    hue="category",
    diag_kind="hist",
    plot_kws={"s": 80, "alpha": 0.8},
)



plt.suptitle("Pairwise scatter of centroid coordinates", y=1.03)
plt.tight_layout()
plt.show()

embeds = centroid_embedding[["gender", "status", "age"]]
from isometry_pursuit.experiment import run_resampling_experiment
from isometry_pursuit.transformation import exponential_transformation
from isometry_pursuit.algorithm import greedy, brute, group_basis_pursuit
from isometry_pursuit.loss import isometry_loss, group_lasso_norm


# Parameters for the experiment
R = 25  # Number of replicates
frac = 0.5  # Fraction of features sampled
power = 1.0  # Experiment power
compute_brute = False  # Placeholder for brute computation flag

# Run the resampling experiment
results = run_resampling_experiment(embeds, 10, frac=frac, R=R, power=power, compute_brute=compute_brute)


co_occurrence_matrix = results[1]

import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(6, 5))

sns.heatmap(
    co_occurrence_matrix,
    ax=ax,
    cmap="viridis",         # any palette you like
    xticklabels=words,      # or e.g. lda_embedding.index
    yticklabels=words,
)

ax.set_xlabel("Word $i$")    # x‑axis label
ax.set_ylabel("Word $j$")    # y‑axis label
ax.set_title("Word‑by‑word co‑occurrence heat‑map")  # overall title

plt.tight_layout()
plt.show()


import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Build the graph exactly as before --------------------------------------
G = nx.Graph()
for word in words:
    G.add_node(word)

# choose a sensible cut‑off: 75th percentile of all non‑zero counts


for i in range(len(words)):
    for j in range(i + 1, len(words)):
        w = co_occurrence_matrix[i, j]
        if w >= 1:
            G.add_edge(words[i], words[j], weight=w)

# focus on the largest connected block
G_sub = G.subgraph(max(nx.connected_components(G), key=len)).copy()


# --- 1.  node sizes  ---------------------------------------------------------
node_strength = co_occurrence_matrix.sum(axis=1)        # total co‑occurrence per word
strength_map = {words[i]: node_strength[i] for i in range(len(words))}

# rescale to a nice visual range
s_min, s_max = 400, 3000
values = np.array(list(strength_map.values()), dtype=float)
sizes = s_min + (values - values.min()) / (values.ptp() + 1e-9) * (s_max - s_min)

# --- 2.  edge widths  --------------------------------------------------------
edge_weights = np.array([d["weight"] for _, _, d in G_sub.edges(data=True)], dtype=float)

# --- node sizes --------------------------------------------------------------
node_strength = co_occurrence_matrix.sum(axis=1)           # total strength per word

# visual range
s_min, s_max = 400, 3000
sizes_all = s_min + (node_strength - node_strength.min()) / (node_strength.ptp()+1e-9) * (s_max - s_min)

# word → size dictionary
size_dict = {word: size for word, size in zip(words, sizes_all)}

# sizes just for the nodes in the subgraph, in the order NetworkX will draw them
node_sizes = [size_dict[n] for n in G_sub.nodes()]

# --- edge widths (unchanged) --------------------------------------------------
edge_weights = np.array([d["weight"] for _, _, d in G_sub.edges(data=True)], dtype=float)
ew_min, ew_max = 0.5, 10
edge_widths = ew_min + (edge_weights - edge_weights.min()) / (edge_weights.ptp()+1e-9) * (ew_max - ew_min)

# --- draw ---------------------------------------------------------------------
pos = nx.spring_layout(G_sub, seed=42, k=0.5)

# -----------------------------------------------------------------
# 0.  map every word -> axis category -----------------------------
axis_of_category = {
    "man": "gender", "woman": "gender",
    "regent": "status", "slave": "status",
    "young": "age", "old": "age",
}

axis_palette = {
    "gender": "tab:purple",
    "status": "tab:orange",    # “power / status”
    "age":    "tab:green",
}

node_colors = [
    axis_palette[axis_of_category[category_of[n]]]  # colour by axis, not sub‑class
    for n in G_sub.nodes()
]

# -----------------------------------------------------------------
# 1.  draw ---------------------------------------------------------
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(
    G_sub,
    pos,
    node_size=node_sizes,
    node_color=node_colors,
    alpha=0.85,
    linewidths=1,
    edgecolors="navy",
)
nx.draw_networkx_edges(
    G_sub,
    pos,
    width=edge_widths,
    alpha=0.7,
    edge_color="steelblue",
)
nx.draw_networkx_labels(
    G_sub,
    pos,
    font_size=9,
    font_weight="bold",
    font_color="black",
)

# legend keyed by axis
from matplotlib.patches import Patch
handles = [Patch(color=c, label=ax) for ax, c in axis_palette.items()]
plt.legend(handles=handles, title="Axis category", loc="upper right")

plt.title("Force‑Directed Graph of Word Co‑occurrences", fontsize=14)
plt.axis("off")
plt.tight_layout()
plt.show()

#greedy
co_occurrence_matrix = results[2]

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# --- Build the graph exactly as before --------------------------------------
G = nx.Graph()
for word in words:
    G.add_node(word)

# choose a sensible cut‑off: 75th percentile of all non‑zero counts


for i in range(len(words)):
    for j in range(i + 1, len(words)):
        w = co_occurrence_matrix[i, j]
        if w >= 1:
            G.add_edge(words[i], words[j], weight=w)

# focus on the largest connected block
G_sub = G.subgraph(max(nx.connected_components(G), key=len)).copy()


# --- 1.  node sizes  ---------------------------------------------------------
node_strength = greedy_co_occurrence_matrix.sum(axis=1)        # total co‑occurrence per word
strength_map = {words[i]: node_strength[i] for i in range(len(words))}

# rescale to a nice visual range
s_min, s_max = 400, 3000
values = np.array(list(strength_map.values()), dtype=float)
sizes = s_min + (values - values.min()) / (values.ptp() + 1e-9) * (s_max - s_min)

# --- 2.  edge widths  --------------------------------------------------------
edge_weights = np.array([d["weight"] for _, _, d in G_sub.edges(data=True)], dtype=float)

# --- node sizes --------------------------------------------------------------
node_strength = co_occurrence_matrix.sum(axis=1)           # total strength per word

# visual range
s_min, s_max = 400, 3000
sizes_all = s_min + (node_strength - node_strength.min()) / (node_strength.ptp()+1e-9) * (s_max - s_min)

# word → size dictionary
size_dict = {word: size for word, size in zip(words, sizes_all)}

# sizes just for the nodes in the subgraph, in the order NetworkX will draw them
node_sizes = [size_dict[n] for n in G_sub.nodes()]

# --- edge widths (unchanged) --------------------------------------------------
edge_weights = np.array([d["weight"] for _, _, d in G_sub.edges(data=True)], dtype=float)
ew_min, ew_max = 0.5, 10
edge_widths = ew_min + (edge_weights - edge_weights.min()) / (edge_weights.ptp()+1e-9) * (ew_max - ew_min)

# --- draw ---------------------------------------------------------------------
pos = nx.spring_layout(G_sub, seed=42, k=0.5)

# -----------------------------------------------------------------
# 0.  map every word -> axis category -----------------------------
axis_of_category = {
    "man": "gender", "woman": "gender",
    "regent": "status", "slave": "status",
    "young": "age", "old": "age",
}

axis_palette = {
    "gender": "tab:purple",
    "status": "tab:orange",    # “power / status”
    "age":    "tab:green",
}

node_colors = [
    axis_palette[axis_of_category[category_of[n]]]  # colour by axis, not sub‑class
    for n in G_sub.nodes()
]

# -----------------------------------------------------------------
# 1.  draw ---------------------------------------------------------
plt.figure(figsize=(8, 8))
nx.draw_networkx_nodes(
    G_sub,
    pos,
    node_size=node_sizes,
    node_color=node_colors,
    alpha=0.85,
    linewidths=1,
    edgecolors="navy",
)
nx.draw_networkx_edges(
    G_sub,
    pos,
    width=edge_widths,
    alpha=0.7,
    edge_color="steelblue",
)
nx.draw_networkx_labels(
    G_sub,
    pos,
    font_size=9,
    font_weight="bold",
    font_color="black",
)

# legend keyed by axis
from matplotlib.patches import Patch
handles = [Patch(color=c, label=ax) for ax, c in axis_palette.items()]
plt.legend(handles=handles, title="Axis category", loc="upper right")

plt.title("Force‑Directed Graph of Word Co‑occurrences", fontsize=14)
plt.axis("off")
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Settings
powers = np.logspace(-1, 1, 5)
frac = 0.5
D_wines = [4, 5, 6]
name = "wine_standardized"
base_path = "/Users/samsonkoelle/isometry-pursuit/experiments"

# Colors
turquoise = "turquoise"  # losses_1 < losses_2 (better)
pink = "pink"       # losses_1 > losses_2 (worse)
gray = "#4E2A84"       # losses_1 ≈ losses_2 (tie)

# Global font size settings
plt.rcParams.update({
    'font.size': 18,
    'axes.titlesize': 24,
    'axes.labelsize': 24,
    'xtick.labelsize': 20,
    'ytick.labelsize': 20,
    'legend.fontsize': 18,
    'figure.titlesize': 28
})

# Initialize figure
fig, axes = plt.subplots(1, len(D_wines), figsize=(7 * len(D_wines), 6), sharey=True)
tolerance = 1e-14

for i, D in enumerate(D_wines):
    ax = axes[i]
    props_better, props_equal, props_worse = [], [], []

    for power in powers:
        power_str = str(power).replace(".", "p")
        frac_str = str(frac).replace(".", "p")
        filename = f"{name}_{D}_{frac_str}_{power_str}"
        filepath = os.path.join(base_path, filename)

        try:
            df = pd.read_csv(filepath)
            losses_1 = df['Losses'].values
            losses_2 = df['Two-Stage Losses'].values
            diff = losses_1 - losses_2

            # Logic: lower loss is better
            props_better.append(np.mean(diff < -tolerance))  # losses_1 better
            props_equal.append(np.mean(np.abs(diff) <= tolerance))  # tie
            props_worse.append(np.mean(diff > tolerance))  # losses_1 worse

        except Exception as e:
            print(f"Failed for {filename}: {e}")
            continue

    xvals = powers[:len(props_better)]
    stacked = [props_better, props_equal, props_worse]

    ax.stackplot(
        xvals, stacked,
        colors=[turquoise, gray, pink],
        linewidth=2,
        edgecolor="black",
        labels=[r"$? = \leq$", r"$=$", r"$\geq$"],
    )
    ax.set_xscale("log")
    ax.set_ylim(0, 1)
    ax.set_xlim(0.1, 10)
    ax.set_title(f"$D = {D}$", fontsize=28)
    ax.set_xlabel("$c$", fontsize=26)
    if i == 0:
        ax.set_ylabel(r"$\widehat P(\ell_c(X_{.\widehat{S}_g}) \; ? \; \ell_c(X_{.\widehat{S}_{TSIP}}))$", fontsize=26)
    ax.grid(True)


# Legend
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=3, frameon=False, bbox_to_anchor=(0.5, 1.05), fontsize = 30)

# Layout
plt.tight_layout()
plt.subplots_adjust(top=0.82)
plt.savefig("/Users/samsonkoelle/isometry-pursuit/figures/grid_power_comparison_filled.png", dpi=300)
plt.show()
```
