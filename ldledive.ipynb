```python
# ??buml_obj.fit
from pyLDLE2 import local_views_
from pyLDLE2.util_ import print_log, sparse_matrix, nearest_neighbors
```

```python
data = X
self = buml_obj
neigh_dist, neigh_ind = nearest_neighbors(data,
                                              k_nn=self.local_opts['k_nn0'],
                                              metric=self.local_opts['metric0'])




d_e = sparse_matrix(neigh_ind, neigh_dist)
d_e = d_e.maximum(d_e.transpose())

# to compute far off points
d_e_small = sparse_matrix(neigh_ind[:,:self.local_opts['k']],
                          neigh_dist[:,:self.local_opts['k']])
d_e_small = d_e_small.maximum(d_e_small.transpose())

neigh_ind = neigh_ind[:,:self.local_opts['k_nn']]
neigh_dist = neigh_dist[:,:self.local_opts['k_nn']]


# Construct low dimensional local views
LocalViews = local_views_.LocalViews(self.exit_at, self.verbose, self.debug)
```

```python
# LocalViews.fit(self.d, data, d_e, neigh_dist, neigh_ind, ddX, self.local_opts)
from pyLDLE2 import gl_, ipge_
```

```python
# def fit(self, , X, d_e, neigh_dist, neigh_ind, ddX, local_opts):
d = buml_obj.d
X = data
local_opts = buml_obj.local_opts

GL = gl_.GL(debug=self.debug)
GL.fit(neigh_dist, neigh_ind, local_opts)

epsilon = neigh_dist[:,[local_opts['k']-1]]
U = sparse_matrix(neigh_ind[:,:local_opts['k']],
                  np.ones((neigh_ind.shape[0],local_opts['k']), dtype=bool))

IPGE = ipge_.IPGE(debug=self.debug)
IPGE.fem(opts = {'phi': GL.phi, 'neigh_ind': neigh_ind, 'k': local_opts['k'],
                     'neigh_dist': neigh_dist, 'epsilon': epsilon,
                     'd': d, 'p': local_opts['p']})


# Compute gamma
gamma = 1/(np.sqrt(U.dot(GL.phi**2)/local_opts['k'])+1e-12)
```

```python
# really its a greedy algorithm
# find most isometric embedding in dimension 1
# the find most isometric in next dimension
# then next most isometric given the last two (less complex but can have bad results)
```
